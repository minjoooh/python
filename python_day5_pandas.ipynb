## 268 pandas로 인구 구조 분석하기

```python
df = pd.read_csv('age.csv', index_col=0, encoding='cp949' )
df.shape
df = df.dropna()
df = df.replace(',', '', regex=True)
a = df.astype('int')
```

```python
a.dtypes
a.info()
```

```python
# 1. 인구를 총 인구수로 나눈다
b = a.div(a.iloc[:, 0], axis=0)
c = b.iloc[:, 2:]
c
```

```python
# 2. 궁금한 지역 이름 입력받는다.
name = input('지역을 입력해 주세요 : ')
```

```python
mask = c.index.str.contains(name)
df2 = c[mask]
df2
```

```python
# 3. 그래프 그리기
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family = 'Malgun Gothic')
df2.T.plot()
```

```python
df = c
x = df.sub(df2.iloc[0], axis=1) # 원본과 입력한 지역의 0번째 행을 뺀다.
y = np.power(x, 2)
z = y.sum(axis=1) # 나이 차이를 다 누적시킨다.
i = z.sort_values().index[:10]
df.loc[i].T.plot()
```

# 야구 데이터 활용하기 (http://www.kbreport.com/main)

```python
url = 'file:///C:/Users/wtime/Downloads/KBReport.html'
tables = pd.read_html(url, header=0, index_col=1)
```

```python
baseball = tables[0]
data = baseball.iloc[:-1]
data.info()
```

```python
# object 형변환
data.loc[:,'승률'] = data['승률'].astype('float')
data.loc[:,'기대승률'] = data['기대승률'].astype('float')
```

```python
# 승률을 bar 그래프로 표현하기
plt.rc('font', family = 'Malgun Gothic')
plt.barh(data.index, data['승률']/data['승률'].max(), label='홈런')
plt.barh(data.index, data['홈런']/data['홈런'].max(), alpha=0.5, label='승률')
plt.legend()
```

- 나의 홈팀과 가장 유사한 성격의 팀을 찾아라

```python
plt.plot(data.index, data['득점'])
plt.scatter(data.index, data['득점'])
plt.bar(data.index, data['득점'])
```

```python
# 승률, 기대승률, 타율, 출루율, 장타율
col = ['승률', '기대승률', '타율', '출루율', '장타율']
for c in col:
    plt.plot(data.index, data[c], label=c)
plt.plot(data.index, data['볼넷%']/data['볼넷%'].max(), label='볼넷')
plt.plot(data.index, data['삼진%']/data['삼진%'].max(), label='삼진')
plt.legend()
```

```python
plt.title('홈런 분포') 
p = plt.pie(data.홈런, labels=data.index, autopct = '%.1f')
```

```python
plt.title('경기당 득점 분포')
plt.hist(data.득점/data.경기, bins=10)
```
