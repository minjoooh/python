```python
# 6월 19일의 최고기온의 히스토그램
f = open('seoul.csv') #29
data = csv.reader(f)
next(data)

date_06_19 = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '06':
            if row[0].split('-')[2] == '19': 
                date_06_19.append(float(row[-1]))
            
```

```python
n = plt.hist(date_06_19, bins=100)
```

```python
# 1월과 8월의 최고기온 히스토그램
f = open('seoul.csv') #29
data = csv.reader(f)
x = np.array(next(data))

jan = []
aug = []
for row in data:
    if row[-1] != '':
        date = row[0].split('-')
        if date[1] == '01':
            jan.append(float(row[-1]))
        if date[1] == '08':
            aug.append(float(row[-1]))            
```

```python
#plt.figure(figsize=(3,1.2))
jan_hist = plt.hist(jan, bins=100, label = 'Jan')
aug_hist = plt.hist(aug, bins=100, label = 'Aug')
plt.legend()
```

```python
# 1 ~ 12 의 최고기온을 박스 그림으로 표현
import pandas as pd
data = pd.read_csv('seoul.csv', encoding='cp949')
data = data.dropna() # 중간에 누락되어 있는 것 제거
data.head() # 상위 5개
```

```python
month = []
for i in range(12):
    month.append([])
    
x = np.array(data)
x.shape, x[0]
```

```python
np.mean(x[:, -1])
```

```python
for i in range(len(x)):
    row = x[i]
    date = row[0]
    high = row[-1]
    mon = date.split('-')[1]
    month[int(mon)-1].append(high) # 1월은 month array에서 0번째 위치에 저장해야하므로
    print(mon, high)
```

```python
weather_12month_plot = plt.boxplot(month)
```
